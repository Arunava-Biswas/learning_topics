```python
import pandas as pd
```

## How to sort a pandas DataFrame or a Series?

```python
# read a dataset of top-rated IMDb movies into a DataFrame

movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

**Note:** None of the sorting methods below affect the underlying data. (In other words, the sorting is temporary).

```python
# sort the 'title' Series in ascending order (returns a Series)
# here we see the first 10 results

movies.title.sort_values().head(10)
```

```python
# sort in descending order instead
# here we need to pass the argument 'ascending=False'

movies.title.sort_values(ascending=False).head(10)
```

```python
# here it returns a series

type(movies.title.sort_values().head(10))
```

```python
# sort the entire DataFrame by the 'title' Series (returns a DataFrame)
# here we use the series name to sort the dataframe

movies.sort_values('title').head(10)
```

```python
# here it is a dataframe

type(movies.sort_values('title').head(10))
```

```python
# sort in descending order on the basis of 'duration'

movies.sort_values('duration', ascending=False).head(10)
```

```python
# sorting on the basis of multiple columns
# sort the DataFrame first by 'content_rating', then by 'duration'

movies.sort_values(['content_rating', 'duration']).head(10)
```
